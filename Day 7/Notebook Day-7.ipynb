```python
dbutils.widgets.text("source_path", "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv")
dbutils.widgets.dropdown("layer", "bronze", ["bronze","silver","gold"])
```

```python

# 1) Read
source = dbutils.widgets.get("source_path")
layer = dbutils.widgets.get("layer")

# 2) Map
NOTEBOOKS = {
    "bronze": "/Workspace/Users/meghanadhage@gmail.com/14 Days AI Challenge/Jobs/bronze_data_ingestion_notebook",
    "silver": "/Workspace/Users/meghanadhage@gmail.com/14 Days AI Challenge/Jobs/silver_data_transformation_notebook",
    "gold":   "/Workspace/Users/meghanadhage@gmail.com/14 Days AI Challenge/Jobs/gold_business_aggregation_notebook"
}

# 3) Runner
def run_layer(layer_name: str, source_path: str, timeout_sec: int = 0) -> str:
    if layer_name not in NOTEBOOKS:
        raise ValueError(f"Unknown layer '{layer_name}'.")
    return dbutils.notebook.run(NOTEBOOKS[layer_name], timeout_seconds=timeout_sec, arguments={"source_path": source_path})

# 4) Orchestrate
def run_pipeline(layer_selector: str, source_path: str):
    if layer_selector == "all":
        for l in ["bronze", "silver", "gold"]:
            res = run_layer(l, source_path)
            print(f"{l} completed: {res}")
    else:
        res = run_layer(layer_selector, source_path)
        print(f"{layer_selector} completed: {res}")

# 5) Execute with basic error handling
try:
    run_pipeline(layer, source)
    dbutils.notebook.exit("OK")
except Exception as e:
    dbutils.notebook.exit(f"FAILED: {str(e)}")
```
